```python
import numpy as np
from matplotlib import pyplot as plt

# Generar datos aleatorios
ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

# Trazar el gráfico de línea con color azul ('b')
plt.plot(x, ys, '-', color='blue')

# Rellenar el área debajo de la línea con color amarillo ('yellow')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='yellow', alpha=0.6)

# Agregar un título al gráfico
plt.title("Ejemplo de Rellenos y Alfa")

# Mostrar el gráfico
plt.show()
```
